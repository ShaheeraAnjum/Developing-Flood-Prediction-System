# Epic 1 & 2: Data Collection and Analysis
This notebook covers downloading, loading, and analyzing the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load dataset
df = pd.read_csv('../data/flood_dataset.csv')
df.head()
```

```python
# Descriptive Statistics
df.info()
df.describe()
```

```python
# Univariate Analysis - Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['Annual Rainfall'], kde=True)
plt.title('Distribution of Annual Rainfall')
plt.show()
```

```python
# Univariate Analysis - Box plots (Outliers)
plt.figure(figsize=(10,6))
sns.boxplot(x=df['Annual Rainfall'])
plt.title('Boxplot of Annual Rainfall')
plt.show()
```

```python
# Multivariate Analysis - Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
